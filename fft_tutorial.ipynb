```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# check FFT performance
x = np.linspace(0, 4*np.pi, 100)
y = np.sin(x)
plt.plot(x,y)
plt.show()
```

```python
# check FFT performance
x = np.linspace(0, 4*np.pi, 100)
y = np.sin(2*x)
plt.plot(x,y)
plt.show()
```

```python
# check FFT performance
x = np.linspace(0, 4*np.pi, 100)
y = np.sin(x) + np.sin(2*x)
plt.plot(x,y)
plt.show()
```

```python
# check FFT performance
x = np.linspace(0, 4*np.pi, 100)
y = np.sin(x) - np.sin(2*x) + np.sin(3*x)
plt.plot(x,y)
plt.show()
```

```python
fft_result = np.fft.fft(y)
absolute = np.abs(fft_result)[:int(len(fft_result)/2)+1]
```

```python
plt.plot(absolute)
plt.show()
```
